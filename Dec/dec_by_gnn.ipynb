```python
import numpy as np
import tensorflow as tf
import time
import os
import sys
from utils import Logger, pcm2gm
from ewgnn_7 import GNN_Decoder, train_model_gnn, test_model_gnn
from bp_7 import E2EModel
```

```python
time_now = time.strftime('%m%d%H%M%S', time.localtime(time.time()))
os.makedirs('./gnn_dec_result/'+time_now)

# log
logfile = open('./gnn_dec_result/'+time_now+'/log.txt', 'a')
sys.stdout = Logger('./gnn_dec_result/'+time_now+'/log.txt', sys.stdout)
```

```python
n = 32
k = 16
code = 'drl'

file_h = './Logs/1106074917/mat/episode_178step_1reward_10.555698805138144.txt'
pcm = np.loadtxt(file_h, dtype=np.int32)
gnm = pcm2gm(pcm, verify_results=True)

num_iter_index = 8
multiloss_flag = True

tf.random.set_seed(2)
bp_decoder = GNN_Decoder(pcm = pcm,
                         num_iter = num_iter_index,
                         multiloss = multiloss_flag)
e2e_model = E2EModel(gnm, bp_decoder)

print('Code: ', n, ',', k, code, 'from', file_h)
print('Iteration: ', num_iter_index)
print('Multi-loss: ', multiloss_flag)

e2e_model(1, 1.) # call model once to init
e2e_model.summary()
```

# Training

```python
train_params = [
    #batch_size, learning_rate, num_iter
    [1.0, 8.0, 4000, 1e-3, 3000]
]
print('Train params:', train_params, '\n')
train_model_gnn(e2e_model, train_params)
e2e_model.save_weights('./gnn_dec_result/'+time_now + '/model_mloss_1')
bers, blers = test_model_gnn(e2e_model, [1.0, 8.0, 100000, 10000, 500])
# print('BER:', bers)
# print('BLER:', blers)

train_params = [
    #batch_size, learning_rate, num_iter
    [1.0, 8.0, 4000, 1e-4, 10000]
]
print('Train params:', train_params, '\n')
train_model_gnn(e2e_model, train_params)
e2e_model.save_weights('./gnn_dec_result/'+time_now + '/model_mloss_2')
bers, blers = test_model_gnn(e2e_model, [1.0, 8.0, 100000, 10000, 500])
# print('BER:', bers)
# print('BLER:', blers)

train_params = [
    #batch_size, learning_rate, num_iter
    [1.0, 8.0, 4000, 1e-4, 10000]
]
print('Train params:', train_params, '\n')
train_model_gnn(e2e_model, train_params)
e2e_model.save_weights('./gnn_dec_result/'+time_now + '/model_mloss_3')
bers, blers = test_model_gnn(e2e_model, [1.0, 8.0, 100000, 10000, 500])
# print('BER:', bers)
# print('BLER:', blers)

train_params = [
    #batch_size, learning_rate, num_iter
    [1.0, 8.0, 4000, 1e-5, 10000]
]
print('Train params:', train_params, '\n')
train_model_gnn(e2e_model, train_params)
e2e_model.save_weights('./gnn_dec_result/'+time_now + '/model_mloss_4')
bers, blers = test_model_gnn(e2e_model, [1.0, 8.0, 100000, 10000, 500])
print('BER:', bers)
print('BLER:', blers)
```

# Testing

```python
n_test = 32
k_test = 16
code = 'drl'

file_h_test = './Logs/1106074917/mat/episode_178step_1reward_10.555698805138144.txt'
pcm_test = np.loadtxt(file_h_test, dtype=np.int32)
gnm_test = pcm2gm(pcm_test, verify_results=True)

num_iter_index_test = 8
# multiloss_flag = True

tf.random.set_seed(2)
bp_decoder_test = GNN_Decoder(pcm = pcm_test, num_iter = num_iter_index_test)
e2e_model_test = E2EModel(gnm_test, bp_decoder_test)

model_fp = './gnn_dec_result/'+time_now + '/model_mloss_4'
e2e_model_test.load_weights(model_fp)

print('Load model from ', model_fp)
print('Test codes', n_test, k_test, code, 'from', file_h_test)
print('Test iteration: ', num_iter_index_test)

test_model_gnn(e2e_model_test, [1.0, 8.0, 10000, 10000, 500])
```
